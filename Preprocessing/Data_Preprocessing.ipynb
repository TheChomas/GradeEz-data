```python

data = next(iter(uploaded.values()))
d = json.loads(data.decode())
```

```python
res2=[]

for k in range(len(d['data'])):
  
  for i in range(len(d['data'][k]['paragraphs'])):
    que=[]
    ans=[]
    pas=[]
    max=0
   
    pas.append(d['data'][k]['paragraphs'][i]['context'])
    for j in range(len(d['data'][k]['paragraphs'][i]['qas'])):
      que.append(d['data'][k]['paragraphs'][i]['qas'][j]['question'])
     
      try:
        ans.append(d['data'][k]['paragraphs'][i]['qas'][j]['answers'][0]['text'])
      except:
        ans.append(d['data'][k]['paragraphs'][i]['qas'][j]['plausible_answers'][0]['text'])
        continue
    db ={
      "paragraph" : d['data'][k]['paragraphs'][i]['context'],
      "question" :que,
      "answers":ans
    }
    res2.append(db)  
```
